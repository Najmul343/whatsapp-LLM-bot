```python
# ✅ Step 1: Install packages
!pip install -q google-generativeai PyPDF2

# ✅ Step 2: Imports
import google.generativeai as genai
import PyPDF2
from google.colab import files
from IPython.display import display, Markdown
import re

# ✅ Step 3: Configure Gemini
genai.configure(api_key="AIzaSyDmAVKXQoVRvqhihUB6u__I945VtfWR67E")  # Replace with your key
model = genai.GenerativeModel("gemini-1.5-flash")

# ✅ Step 4: Upload Resume
print("📄 Upload your resume (PDF):")
uploaded = files.upload()

resume_text = ""
filename = ""

for fname in uploaded:
    filename = fname
    with open(fname, "rb") as f:
        reader = PyPDF2.PdfReader(f)
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
                resume_text += page_text.strip()

if not resume_text:
    print("❌ Could not extract text. Maybe it's a scanned resume.")
else:
    print("✅ Resume text extracted successfully.")

# ✅ Step 5: Paste Job Description
print("\n📋 Paste your entire job description below (single input):")
job_description = input("Paste JD here:\n").strip()

if not job_description:
    print("❌ No job description provided.")
else:
    # ✅ Step 6: Run Gemini Analysis
    prompt = f"""
Compare the resume and job description.

1. Give a **Fit Score (0–100)**.
2. List **matching strengths**.
3. List **missing skills/gaps**.
4. Recommend improvements.

RESUME:
{resume_text[:12000]}

JOB DESCRIPTION:
{job_description[:5000]}
"""

    try:
        print("\n🧠 Gemini is analyzing...\n")
        response = model.generate_content(prompt)
        reply = response.text.strip()

        # ✅ Extract score
        match = re.search(r"(\d{1,3})", reply)
        score = int(match.group(1)) if match else None

        print("🎯 Fit Score:", score if score is not None else "Not Found", "/ 100\n")
        display(Markdown("### 📊 Fit Report:\n" + reply))

        # ✅ Save result
        result_filename = f"fit_report_{filename.replace('.pdf', '')}.txt"
        with open(result_filename, "w", encoding="utf-8") as f:
            f.write("📄 Resume File: " + filename + "\n")
            f.write("🎯 Fit Score: " + str(score) + " / 100\n\n")
            f.write("📊 Report:\n" + reply)

        print(f"\n📥 Fit report saved as: {result_filename}")
        files.download(result_filename)

    except Exception as e:
        print("❌ Gemini error:", e)
```

```python
# ✅ Step 2: Save the Streamlit app to app.py
%%writefile app.py
import streamlit as st
import google.generativeai as genai
import PyPDF2
import re
import os

# ✅ Gemini API Key
genai.configure(api_key="AIzaSyDmAVKXQoVRvqhihUB6u__I945VtfWR67E")  # Replace with your API key
model = genai.GenerativeModel("gemini-1.5-flash")

st.set_page_config(page_title="Gemini Resume Analyzer", page_icon="📄")
st.title("📄 Resume vs JD Fit Checker + Rewriter (Gemini 1.5 Flash)")
st.markdown("Upload your resume, paste any job description, and get a fit score, suggestions, and rewritten CV.")

resume_file = st.file_uploader("Upload your Resume (PDF)", type=["pdf"])
job_desc = st.text_area("Paste Job Description", height=200, placeholder="e.g. Python Developer with FastAPI, Docker, AWS...")

# ✅ Extract resume text from PDF
def extract_text_from_pdf(file):
    reader = PyPDF2.PdfReader(file)
    text = ""
    for page in reader.pages:
        content = page.extract_text()
        if content:
            text += content.strip() + "\n"
    return text[:15000]

# ✅ Gemini Fit Scoring Function
def get_fit_score(resume_text, jd_text):
    prompt = f"""
Compare the following resume and job description:
1. Give a **Fit Score (0–100)**.
2. List **matching skills/experience**.
3. List **missing skills/gaps**.
4. Recommend **resume improvements**.

RESUME:
{resume_text}

JOB DESCRIPTION:
{jd_text}
"""
    try:
        response = model.generate_content(prompt)
        reply = response.text.strip()
        match = re.search(r"(\d{1,3})", reply)
        score = int(match.group(1)) if match else 0
        score = max(0, min(score, 100))  # Clamp score between 0–100
        return score, reply
    except Exception as e:
        return 0, f"❌ Error from Gemini: {e}"

# ✅ Resume Rewriter Function
def rewrite_resume(resume_text, jd_text):
    prompt = f"""
You are an AI resume assistant. Rewrite the resume below to better match the job description.
Make it professional, ATS-friendly, and role-focused.

RESUME:
{resume_text}

JOB DESCRIPTION:
{jd_text}
"""
    try:
        response = model.generate_content(prompt)
        return response.text.strip()
    except Exception as e:
        return f"❌ Error rewriting resume: {e}"

# ✅ App Logic
if st.button("🔍 Analyze Fit"):
    if not resume_file:
        st.warning("⚠️ Please upload your resume.")
    elif not job_desc.strip():
        st.warning("⚠️ Please paste a job description.")
    else:
        with st.spinner("Analyzing with Gemini..."):
            resume_text = extract_text_from_pdf(resume_file)
            score, feedback = get_fit_score(resume_text, job_desc)

        st.success(f"🎯 Fit Score: {score} / 100")
        st.markdown("### 💬 Gemini Suggestions:")
        st.markdown(feedback)
        st.slider("📈 Visual Fit Score", min_value=0, max_value=100, value=score, disabled=True)

        filename = f"fit_report.txt"
        with open(filename, "w", encoding="utf-8") as f:
            f.write(f"Resume: {resume_file.name}\n")
            f.write(f"Fit Score: {score}/100\n\n")
            f.write(feedback)

        with open(filename, "rb") as f:
            st.download_button("📥 Download Fit Report", f, file_name=filename, mime="text/plain")

if st.button("✍️ Rewrite Resume"):
    if not resume_file or not job_desc.strip():
        st.warning("⚠️ Upload resume and paste JD first.")
    else:
        with st.spinner("Rewriting resume with Gemini..."):
            resume_text = extract_text_from_pdf(resume_file)
            rewritten = rewrite_resume(resume_text, job_desc)
        st.markdown("### 📝 Rewritten Resume:")
        st.markdown(rewritten)
        with open("rewritten_resume.txt", "w", encoding="utf-8") as f:
            f.write(rewritten)
        with open("rewritten_resume.txt", "rb") as f:
            st.download_button("📄 Download Rewritten Resume", f, file_name="rewritten_resume.txt", mime="text/plain")
```
